```python
#the node class which contain the matrix stored at each node its depth cost and the node which created it
#this is used to produce a path the secondary depth huristic used in the A* algorithm
class Node:
    def __init__(self, current, depth, cost):
        self.currentMatrix = current
        self.depth = depth
        self.cost = cost
        self.pastnode = None
    def getdepth():
        return(depth)
    def getCurrentState():
        return(self.currentMatrix)
```
